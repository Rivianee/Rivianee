```python
import streamlit as st
import pandas as pd
from lightgbm import LGBMClassifier

# Função para carregar os dados
def load_data():
    df = pd.read_excel('/content/PLANILHA_ATUALIZADA_MODELAGEM_30_05_2024 (1).xlsx')
    # Remover a coluna 'nm_participante'
    df.drop(columns=['nm_participante'], inplace=True)
    # Codificar as variáveis categóricas
    categoric_cols = ['dsc_statusplano2', 'estado_civil', 'dsc_perfilinvestimento', 'Categoria - PROFISSÃO', 'dsc_sexo', 'estado_correspondencia', 'estado']
    encoded_cols = pd.get_dummies(df[categoric_cols], drop_first=True)
    # Concatenar as variáveis codificadas com o DataFrame original
    df_encoded = pd.concat([df.drop(columns=categoric_cols), encoded_cols], axis=1)
    return df_encoded

# Função para carregar e treinar o modelo
def load_model():
    model = LGBMClassifier()
    # Treinar o modelo com todos os dados disponíveis
    df = load_data()
    X = df.drop(columns=['fez_portabilidade'])
    y = df['fez_portabilidade']
    model.fit(X, y)
    return model

# Função principal do aplicativo
def main():
    st.title('Aplicativo de Portabilidade')

    # Carregar os dados e o modelo
    df = load_data()
    model = load_model()

    # Mostrar a imagem no aplicativo principal
    image_path = '/content/WhatsApp Image 2024-05-31 at 13.57.46.jpeg'  # Substitua pelo caminho correto da sua imagem
    try:
        st.image(image_path, caption='Descrição da imagem', use_column_width=True)
    except FileNotFoundError:
        st.warning('Imagem não encontrada. Verifique o caminho e tente novamente.')

    # Mostrar as variáveis importantes
    st.subheader('Variáveis Importantes')
    feature_importances = pd.DataFrame({'Feature': df.drop(columns=['fez_portabilidade']).columns, 'Importance': model.feature_importances_})
    feature_importances = feature_importances.sort_values(by='Importance', ascending=False)
    st.dataframe(feature_importances)

    # Inputs do usuário na barra lateral
    st.sidebar.header('Insira os dados do cliente')
    inputs = {}
    for feature in df.columns:
        if feature != 'fez_portabilidade':
            inputs[feature] = st.sidebar.text_input(feature, value='')

    # Previsão do modelo
    if st.sidebar.button('Prever'):
        input_df = pd.DataFrame(inputs, index=[0])
        prediction = model.predict(input_df)
        st.write('Resultado da Previsão:')
        if prediction[0] == 1:
            st.write('O cliente fará portabilidade.')
        else:
            st.write('O cliente não fará portabilidade.')

    # Aba de texto explicativo
    st.sidebar.header('Informações Adicionais')
    st.sidebar.markdown("""
        ### Sobre o Aplicativo
        Este aplicativo prevê se um cliente fará portabilidade com base nos dados inseridos.
        ### Como usar
        Insira os dados do cliente nos campos fornecidos na barra lateral e clique no botão 'Prever'.
        ### Sobre o Modelo
        O modelo utilizado é um LGBMClassifier treinado com dados históricos para prever a portabilidade.
    """)

if __name__ == "__main__":
    main()
```
